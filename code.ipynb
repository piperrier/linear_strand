## 1. (7, 4) Hamming code (trivial)

```python
R = PolynomialRing(GF(2), 4, 'x', order ="lex");
G = matrix(GF(2),
[[1, 0, 0, 0, 1, 1, 1],
[0, 1, 0, 0, 1, 1, 0],
[0, 0, 1, 0, 1, 0, 0],
[0, 0, 0, 1, 0, 0, 1]])
print(G)
print("")

G2 = carre_matrice(G)
print(G2)
print("")

i2 = ideal2(G,R)
print(i2)
print("")

M23 = macaulay23(i2,R)
print(M23)
print("")

i3 = M23.right_kernel()
print(i3)
print("")

B_gen = i3.matrix()
print(B_gen)
```

## 2. (16, 11) Hamming code (medium)
For some mistery reasons this seems to be really slow, I still don't know why.

```python
R = PolynomialRing(GF(2), 11, 'x', order ="lex");
C = codes.HammingCode(GF(2), 4)
G = C.generator_matrix()
print(f"G\n{G}")
print("")

start1 = time.time()
G2 = carre_matrice(G)
end1 = time.time()
print(f"[{end1-start1}]G2\n{G2}")
print("")

start2 = time.time()
i2 = ideal2(G,R)
end2 = time.time()
print(f"[{end2-start2}]i2\n{i2}")
print("len(i2)",len(i2))
print("")

start3 = time.time()
M23 = macaulay23(i2,R)
end3 = time.time()
#print(f"[{end3-start3}]M23 [{M23.dimensions()}]\n{M23}")
print(f"[{end3-start3}]M23.dim() [{M23.dimensions()}]")
print("")

start4 = time.time()
i3 = M23.right_kernel()
end4 = time.time()
print(f"[{end4-start4}]i3")
print("")

B_gen = i3.matrix()
#print("B_gen",B_gen)
```
